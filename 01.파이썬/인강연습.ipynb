# 11. 파이썬 함수의 활용과 람다 함수

```python
# 11-5
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)
print(factorial(5))
```

```python
fact = lambda x: x == 0 and 1 or x * fact(x-1)
print(fact(5))
```

```python
l =[]
for i in range(2,10):
    for j in range(1,10):
        l.append("{} X {} = {}" .format(i,j,i*j))
print(l)
```

```python
print([(lambda x,y: '{} X {} = {}' .format(x,y,x*y))(x,y) for x in range(2,10) for y in range(1,10)])
```

# 15. 파이썬 클래스의 활용

```python
# 객체 + other간 add 연산 안됨
class A:
    def __init__(self, i):
        self.i =i
n = A(40)
print(n+1)
```

```python
# 객체 + other간 연산은 add 메소드 재정의 
class A:
    def __init__(self, i):
        self.i =i
    def __add__(self, other):
        return self.i + other
n = A(40)
print(n+1)
```

```python
class MyInteger:
    def __init__(self, i):
        self.i = i
    def __str__(self):
        return str(self.i)
    def __add__(self, other):
        return self.i + other
    def __sub__(self, other):
        return self - other
i = MyInteger(10)
print(str(i))
i = i + 10
print(i)
i += 15
print(i)
```

```python
# 연산자 오버로딩을 하지 않으면 인스터스 간 연산이 되지 않음
i = Myinteger(10)
i *= 10
print(i)
```

```python
i = MyInteger(10)
print(i + 10)
print(10 + i) # 왼쪽 피연산자, 오른쪽 객체가 오는경우 __radd__ 로 오버로딩
```

```python
class A:
    def __init__(self, i):
        self.i = i
    def __add__(self, other):
        return A(self.i + other)
i = A(10)
i = i+1
print(i)
print(type(i))
print(isinstance(i,A))
```

```python
# 인스턴스 객체 a는 초기 값으로 10을 생성
# a와 5의 더하기 a+5 -> 10-5=5
class A:
    def __init__(self, i):
        self.i = i
    def __str__(self):
        return str(self.i)
    def __add__(self, other):
        return self.i - other
a = A(10)
print(a + 5)
```

```python
class Asia:
    def __init__(self, name):
        self.name = name
    def show(self):
        return '해당 국가는 아시아에 있습니다'
class Korea(Asia):
    def __init__(self, name, population, capital):
        Asia.__init__(self, name)
        self.population = population
        self.capital = capital
    def show_name(self):
        return '국가 이름은 : ', self.name
a = Asia('대한민국')
print(a.show())
```

```python
b = Korea('대한민국','5천만','서울')
print(b.show_name())
```

```python
# 숫자가 아닌 것이 입력되었을 때 예외 처리
# 9보다 클 때는 9단을 출력
# 1이 입력되었을 땐 2단 출력
num = input("2-9 사이의 숫자를 입력해주세요 : ")
try:
    if(num.isdigit()):
        num = int(num)
    if(num > 9):
        num = 9
    elif(num < 2):
        num = 2
    else:
        num = num
    for i in range(1,10):
        print("{} X {} = {}" .format(num, i, num*i))
except Exception as msg:
    print("예외 발생", msg)
```
