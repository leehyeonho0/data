```python
import csv
f = open('subwaytime_daejon_202012.csv')
data = csv.reader(f)
next(data)
mx = [0] * 24
mx_station = [''] * 24
for row in data :
    row[4:] = map(int, row[4:])
    for j in range(24) :
        
        a = row[j + 4]
        if a > mx[j] :
            mx[j] = a
            mx_station[j] = row[2]+'('+str(j+4)+'시)'  # 추가
print(mx_station)
print(mx)
```
