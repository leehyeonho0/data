```python
# 데이터 읽어오기
import csv
import matplotlib.pyplot as plt
f = open('daejon.csv')
data = csv.reader(f)
next(data) # 제목줄(첫칸) 없애기

for row in data :
    print(row)

f.close()
```

```python
# 데이터의 맨 윗줄의 헤더 부분을 next() 함수를 사용해 제외하고 
# 최고기온 데이터만 출력하기
f = open('daejon.csv')
data = csv.reader(f)
next(data)

for row in data :
    print(row[-1])

f.close() 
```

```python
# 데이터 리스트에 저장하기
import csv
f =open('daejon.csv')
data = csv.reader(f)
next(data)

result = [] # 최고 기온 데이터를 저장할 리스트 생성

for row in data :
    if row[-1] != '':  # 최고 기온 데이터 값이 존재한다면
        result.append(float(row[-1]))  # result 리스트에 최고 기온 값 추가
print(result)
        
f.close()
```

```python
# 개수확인
print(len(result))
```

```python
# 기상 관측 이래 서울의 최고 기온이 가장 높았던 날
import csv
max_temp =-999   # 최고 기온 값을 저장할 변수
max_date =''       # 최고 기온이 가장 높았던 날짜를 저장할 변수

f =open('daejon.csv')
data = csv.reader(f)
header =next(data)

for row in data :
    if row[-1] =='' :
        row[-1] =-999   # -999를 넣어 빈 문자열이 있던 자리라고 표시
    row[-1] = float(row[-1])
    
    if max_temp < row[-1] :
        max_date = row[0]
        max_temp = row[-1]

f.close()
print('기상 관측 이래 대전의 최고 기온이 가장 높았던 날은',max_date+'로, ', max_temp, '도 였습니다.')
```

```python
# 매년 7월 2일의 최고 기온 데이터 추출 그래프로 나타내기

f = open('daejon.csv')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '07' and row[0].split('-')[2] == '02' :
            result.append(float(row[-1]))

plt.plot(result, 'hotpink')
plt.show()
```

```python
# 최고 기온과 최저기온을 함께 나타내기
f = open('daejon.csv')
data = csv.reader(f)
next(data)
high = []
low = []

for row in data :
    if row[-1] != '' and row[-2] != '' :
        if row[0].split('-')[1] == '07' and row[0].split('-')[2] == '02' :
            high.append(float(row[-1]))  # 최고기온 값을 high 리스트에 저장
            low.append(float(row[-2]))  # 최저 기온 값을 low 리스트에 저장

plt.plot(high, 'hotpink')
plt.plot(low, 'skyblue')
plt.show()
```

```python
# plt.rcParams['axes.unicode_minus'] = False 추가

import csv
import matplotlib.pyplot as plt

f = open('daejon.csv')
data = csv.reader(f)
next(data)
high = [] # 최고 기온 값을 저장할 리스트 high 생성
low = [] # 최저 기온 값을 저장할 리스트 low 생성

for row in data :
    if row[-1] != '' and row[-2] != '' : # 최고 기온 값과 최저 기온 값이 존재한다면
        date = row[0].split('-') # 날짜 값을 – 문자를 기준으로 구분하여 저장
        if 1981 <= int(date[0]) : # 1981년 이후 데이터라면
            if date[1] == '07' and date[2] == '02' : # 7월 2일이라면
                high.append(float(row[-1])) # 최고 기온 값을 high 리스트에 저장
                low.append(float(row[-2])) # 최저 기온 값을 low 리스트에 저장

plt.rc('font', family = 'Malgun Gothic') # 맑은 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.title('내 생일의 기온 변화 그래프') # 제목 설정
plt.plot(high, 'hotpink', label = 'high') # high 리스트에 저장된 값을 hotpink 색으로 그리고 레이블을 표시
plt.plot(low, 'skyblue', label = 'low') # low 리스트에 저장된 값을 skyblue 색으로 그리고 레이블을 표시
plt.legend() # 범례 표시
plt.show() # 그래프 나타내기
```
