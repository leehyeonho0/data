```python
mdf = df.copy()
```

```python
mdf.info()
```

```python
display(mdf.Age.mean())
display(mdf.Age.value_counts().head())
```

```python
mdf.Age.fillna(round(mdf.Age.mean(),0), inplace=True)
mdf.info()
```

```python
pd.crosstab([mdf.Sex],mdf.Pclass,margins=True).style.background_gradient(cmap='summer_r')
```

```python
# Pclass_s = mdf.groupby(['Sex', 'Pclass'])['Survived'].aggregate("mean").unstack()
Pclass_s = mdf.pivot_table('Survived', index='Sex', columns='Pclass')
Pclass_s
```

```python
mdf['Pclass_band']=0
mdf.loc[(mdf['Sex']==0)&(mdf['Pclass']==1),'Pclass_band'] = Pclass_s.iloc[0,0]
mdf.loc[(mdf['Sex']==0)&(mdf['Pclass']==2),'Pclass_band'] = Pclass_s.iloc[0,1]
mdf.loc[(mdf['Sex']==0)&(mdf['Pclass']==3),'Pclass_band'] = Pclass_s.iloc[0,2] 
mdf.loc[(mdf['Sex']==1)&(mdf['Pclass']==1),'Pclass_band'] = Pclass_s.iloc[1,0]
mdf.loc[(mdf['Sex']==1)&(mdf['Pclass']==2),'Pclass_band'] = Pclass_s.iloc[1,1]
mdf.loc[(mdf['Sex']==1)&(mdf['Pclass']==3),'Pclass_band'] = Pclass_s.iloc[1,2]
mdf.head(2)
```

```python
corr = mdf.corr()
corr.Survived.sort_values(ascending=False)
```

```python
drop_columns = ['Name','Age','SibSp','Parch','Pclass','Ticket','Embarked']
mdf = mdf.drop(drop_columns, axis=1)
mdf.head()
```

```python
from sklearn.model_selection import train_test_split  
```

```python
dfX = mdf.drop(['PassengerId','Survived'], axis=1)  # exclude 'id' attribute & class variable
dfy = mdf['Survived']                    # class variable
X_train, X_test, y_train, y_test = train_test_split(dfX, dfy, test_size=0.25, random_state=0)
```

```python
display(X_train.shape, X_test.shape)
X_train.head(3)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier(max_depth=6, random_state=0)
```

```python
tree.fit(X_train, y_train)
```

```python
pred_tree = tree.predict(X_test)
pred_tree
```

```python
from sklearn.svm import SVC 
svm = SVC(random_state=0)
svm.fit(X_train, y_train)
```

```python
from sklearn.neural_network import MLPClassifier 
mlp = MLPClassifier()
mlp.fit(X_train, y_train)
```

```python
display(tree.score(X_train, y_train))
display(tree.score(X_test, y_test))
```

```python
display(svm.score(X_train, y_train))
display(svm.score(X_test, y_test))
```

```python
display(mlp.score(X_train, y_train))
display(mlp.score(X_test, y_test))
```

```python
best_model = tree 
best_model.score(X_test, y_test)
```

```python
from sklearn.dummy import DummyClassifier
print(y_test.value_counts())
DummyClassifier(strategy='most_frequent').fit(X_train, y_train).score(X_test, y_test)
```
