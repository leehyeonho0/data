### 타입2

```python
# 파생변수를 저장할 빈 리스트를 만든다
features = []
```

```python
# 파생변수1 : 코너별구매건수
f = df1.groupby('custid')['구매건수'].agg([('코너별구매건수', 'sum')]).reset_index()
features.append(f)
f
```

```python
# 파생변수1 : 코너별구매건수
f = df1.groupby('custid')['총구매액'].agg([('코너별구매액', 'sum')]).reset_index()
features.append(f)
f
```

```python
# 평균구매가격
f = df1.groupby('custid')['총구매액'].agg([('평균구매가격', 'mean')]).reset_index()
features.append(f); f
```

```python
features
```

#### 아래 코드를 수정하면 생성한 모든 파생변수가 병합되고 학습용과 평가용으로 나뉘어 진다.

```python
df_train = pd.DataFrame({'custid': df1.custid.unique()})
df_train
```

```python
for f in features:
    df_train = pd.merge(df_train, f, how='left')
    
display(df_train)    
```
