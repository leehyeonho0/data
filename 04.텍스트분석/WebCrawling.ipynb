##### 원하는 영역에서 하이퍼링크 모두 추출하기

```python
news = 'https://news.daum.net/'
soup = bs(ur.urlopen(news).read(), 'html.parser')
```

```python
# find_all 로 <a> 태그 추출하기
for i in soup.find_all('div', {"class":"item_issue"}):
    print(i.find_all('a'))
```

```python
# get으로 href 속성값 구하기
for i in soup.find_all('div', {"class":"item_issue"}):
    print(i.find_all('a')[0]) # find_all 로 찾아낸 첫 번째 원소를 지정
```

```python
# get으로 href 속성값 구하기
for i in newsoup :
    print(i.find_all('a')[0].get('href'))
# 총 네 개의 URL 주소를 추출 할 수 있다.   
```

##### 기사 제목과 내용을 한꺼번에 추출하기

##### 하이퍼링크된 모든 기사의 제목과 본문 추출하기

```python
# 여러 링크를 사용해 모든 머리기사의 제목과 내용을 가져오기
for i in headline:
    print(i.text, '\n')
```

```python
for i in headline:
    print(i.find_all('a')[0].get('href'))
```

```python
# 제목을 출력할때 마다 해당 기사의 내용을 함께 출력한다.
for i in headline:  # headline 객체에서 <div> 태그를 하나씩 가져온다.
    print(i.text, '\n') # 기사제목을 출력한다.
    
    #  해당 기사가 올라와 있는 웹 사이트를 열어 soup3 객체에 저장한다.    
    soup3 = bs(ur.urlopen(i.find_all('a')[0].get('href')).read(), 'html.parser')  
    
    for j in soup3.find_all('p'): # <p> 태그에서 텍스트만 출력한다.
        print(j.text)
```
