```python
# 제목을 출력할때 마다 해당 기사의 내용을 함께 출력한다.
for i in newsoup:  # headline 객체에서 <div> 태그를 하나씩 가져온다.
    print(i.text, '\n') # 기사제목을 출력한다.
    
    #  해당 기사가 올라와 있는 웹 사이트를 열어 soup 객체에 저장한다.    
    article1 = 'https:' + i.find_all('a')[0].get('href')
    soup = bs(ur.urlopen(article1).read(), 'html.parser')  
    
    for j in soup.find_all('div', {"class":"detail"}): # <p> 태그에서 텍스트만 출력한다.
        print(j.text)
```

```python
import pandas as pd
```

```python
title = []
content = []
# 제목을 출력할때 마다 해당 기사의 내용을 함께 출력한다.
for i in newsoup:  # headline 객체에서 <div> 태그를 하나씩 가져온다.
    title.append(i.text) # 기사제목을 저장한다.
    
    #  해당 기사가 올라와 있는 웹 사이트를 열어 soup 객체에 저장한다.    
    article1 = 'https:' + i.find_all('a')[0].get('href')
    soup = bs(ur.urlopen(article1).read(), 'html.parser')  
    
    for j in soup.find_all('div', {"class":"detail"}): # <p> 태그에서 텍스트만 출력한다.
        content.append(j.text) # 기사 내용을 저장한다.
articles = {'기사제목' : title,
            '기사내용' : content}  
```

```python
mbn_news = pd.DataFrame.from_dict(articles)
mbn_news.head()
```

```python
# 비주얼스튜디오로 코드 저장 (WebCrawling_mbn.py)
import os, re, csv
import requests # URL 주소에 있는 내용을 요청할 때 사용하는 모듈
import urllib.request as ur  # 웹에서 얻은 데이터를 다루는 파이썬 패키지
from bs4 import BeautifulSoup as bs

# 기사 제목, 본문, 하이퍼링크 파일로 저장하기
url = 'https://www.mbn.co.kr/news/date/'
soup = bs(ur.urlopen(url).read(), 'html.parser')

f = open('article_mbn.txt','w')
for i in soup.find_all('dt', {"class":"tit"}):
    try:
        f.write(i.text + '\n')  # 기사 제목
        f.write('http:' + i.find_all('a')[0].get('href') + '\n')  # 기사 URL링크
        soup2 = bs(ur.urlopen('http:' + i.find_all('a')[0].get('href')).read(), 'html.parser')
        for j in soup2.find_all('div', {"class":"detail"}):
            f.write(j.text + '\n')   # 기사 본문
    except:
        pass
f.close()
```

```python
# 비주얼스튜디오로 코드 저장 (WebCrawling_mbn.py)
import os, re, csv
import requests # URL 주소에 있는 내용을 요청할 때 사용하는 모듈
import urllib.request as ur  # 웹에서 얻은 데이터를 다루는 파이썬 패키지
from bs4 import BeautifulSoup as bs

# 기사 제목, 본문, 하이퍼링크 파일로 저장하기
url = 'https://www.mbn.co.kr/news/date/'
soup = bs(ur.urlopen(url).read(), 'html.parser')

f = open('article_mbn.txt','w')
for i in soup.find_all('dt', {"class":"tit"}):
    try:
        f.write('='*60 + '\n')
        f.write(' 기사 제목 : ')
        f.write(i.text + '\n')  # 기사 제목
        
        f.write(' URL : ')
        f.write('http:' + i.find_all('a')[0].get('href') + '\n')  # URL링크
    
        soup2 = bs(ur.urlopen('http:' + i.find_all('a')[0].get('href')).read(), 'html.parser')
        for j in soup2.find_all('div', {"class":"detail"}):
            f.write('\n' + ' 기사 본문 : ')
            f.write(j.text + '\n')   # 기사 본문

    except:
        pass
f.close()
```
